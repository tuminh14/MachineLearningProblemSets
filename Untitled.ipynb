# App

 ## Drop duplicate Row 'App'

```python
(data['App'].value_counts())[data['App'].value_counts()>1]
```

```python
def makeAppDuplicatesName(dataApp,data):
    AppList=data['App'].value_counts()>1
    listIndex=AppList[AppList].index
    listApp=[]
    for i in dataApp:
        listApp.append(i in listIndex)

    return listApp
    
data1=data
data1['isSameOneApp']=makeAppDuplicatesName(data1['App'],data1)
```

```python
data1[data1['isSameOneApp']].sort_values(by='App')
```

```python
data1.drop_duplicates(inplace=True)
```

# Size

```python
data[data['Size'].str.contains('\D\W',regex=True)]['Size'].unique()
```

```python
data[data['Size'].str.contains('\d[kK]',regex=True)]['Size'].unique()
```

```python
def isVariesWithDevice(size):
    if size=='Varies with device':
        return 1
    else :
        return 0
```

```python
data['Size_VariesWithDevice']=data['Size'].apply(isVariesWithDevice)
```

```python
import re
def fixSizeToKB(size):
    if re.search('\d[Mm]',size):
        return float(size[:-1])*1024
    elif  re.search('\d[Kk]',size):
        return float(size[:-1])
```

```python
data['Size']=data['Size'].apply(fixSizeToKB)
```

```python
data.head()
```

# Installs

```python
data[~data['Installs'].str.contains('\d[+]')]
```

```python
def toIntInstalls(install):
    install=re.sub(r'[,]','',install)
    if re.search('\d[+]',install):
        return (int)(install[:-1])
    else:
        return 0
```

```python
data['Installs']=data['Installs'].apply(toIntInstalls)
```
